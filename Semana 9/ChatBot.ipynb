```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Descargar el modelo y el tokenizador
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
import torch
def generate_response(prompt, chat_history_ids=None, max_length=1000):
    new_user_input_ids = tokenizer.encode(prompt + tokenizer.eos_token, return_tensors='pt')
    bot_input_ids = torch.cat([chat_history_ids, new_user_input_ids], dim=-1) if chat_history_ids is not None else new_user_input_ids

    chat_history_ids = model.generate(
        bot_input_ids, max_length=max_length,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_p=0.92,
        top_k=50
    )

    response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
    return response, chat_history_ids
```

```python
def chat():
    chat_history_ids = None
    print("Chatbot: ¡Hola! ¿En qué puedo ayudarte hoy? Escribe 'salir' para terminar.")
    while True:
        user_input = input("Tú: ")
        if user_input.lower() == 'salir':
            print("Chatbot: ¡Adiós!")
            break
        prompt = f"Tú: {user_input}\nChatbot:"
        response, chat_history_ids = generate_response(prompt, chat_history_ids)
        print(f"Chatbot: {response}")

if __name__ == "__main__":
    chat()
```

```python
from sklearn.model_selection import train_test_split

# Si tu dataset no está dividido
train_data, val_data = train_test_split(dataset['train'], test_size=0.1, random_state=42)

# Crear un nuevo dataset con divisiones
dataset = {
    "train": train_data,
    "validation": val_data
}
```

```python
from datasets import Dataset
# Convertir los diccionarios a datasets de Hugging Face
train_dataset = Dataset.from_dict(dataset['train'])
validation_dataset = Dataset.from_dict(dataset['validation'])
```

```python

train_dataset = train_dataset.map(lambda x: {"input_text": x['prompt'], "target_text": x['response']})
validation_dataset = validation_dataset.map(lambda x: {"input_text": x['prompt'], "target_text": x['response']})
```

```python
def tokenize_function(examples):
    # Usa tu tokenizer para procesar las columnas de texto
    model_inputs = tokenizer(examples['input_text'], max_length=512, truncation=True, padding=True)

    # Tokeniza también el texto objetivo
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples['target_text'], max_length=512, truncation=True, padding=True)

    model_inputs['labels'] = labels['input_ids']
    return model_inputs

# Aplicar la tokenización al dataset completo
tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_validation_dataset = validation_dataset.map(tokenize_function, batched=True)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
from transformers import Trainer, TrainingArguments


# Asignar el token de fin de secuencia como el token de relleno
tokenizer.pad_token = tokenizer.eos_token

# # Continuar con la tokenización y el mapeo
# def tokenize_function(examples):
#     model_inputs = tokenizer(examples["input_text"], max_length=32, truncation=True, padding="max_length")
#     labels = tokenizer(examples["target_text"], max_length=32, truncation=True, padding="max_length").input_ids
#     model_inputs["labels"] = labels
#     return model_inputs

# train_dataset = train_dataset.map(lambda x: {"input_text": x['prompt'], "target_text": x['response']})
# validation_dataset = validation_dataset.map(lambda x: {"input_text": x['prompt'], "target_text": x['response']})

def tokenize_function(example):
    # Tokeniza el input_text y target_text
    model_inputs = tokenizer(
        example['input_text'],
        max_length=128,
        truncation=True,
        padding='max_length'
    )

    # Tokeniza el target_text con el mismo tokenizer
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            example['target_text'],
            max_length=128,
            truncation=True,
            padding='max_length'
        )

    model_inputs['labels'] = labels['input_ids']
    return model_inputs

# Aplica la función de tokenización al conjunto de datos
tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_validation_dataset = validation_dataset.map(tokenize_function, batched=True)

# Verifica las columnas después de la tokenización
print(tokenized_train_dataset.column_names)
print(tokenized_validation_dataset.column_names)


```

```python
from sklearn.model_selection import train_test_split

# Elimina las columnas innecesarias del dataset tokenizado
tokenized_train_dataset = tokenized_train_dataset.remove_columns(["response", "prompt", "input_text", "target_text"])
tokenized_validation_dataset = tokenized_validation_dataset.remove_columns(["response", "prompt", "input_text", "target_text"])

# Define los argumentos de entrenamiento
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",  # Realizar la evaluación al final de cada época
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=10,
    weight_decay=0.01,
    logging_dir='./logs',
)

# Crea el Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,  # Conjunto de entrenamiento
    eval_dataset=tokenized_validation_dataset,  # Conjunto de validación (necesario)
    tokenizer=tokenizer,
)

# Entrena el modelo
trainer.train()
```

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Cargar el modelo y el tokenizador preentrenados de GPT-2
model_name = "gpt2-large"  # Puedes usar "gpt2-medium", "gpt2-large", etc., para versiones más grandes del modelo
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Mover el modelo a la GPU si está disponible
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Función para generar respuestas
def generate_response(prompt, max_length=100, num_return_sequences=1):
    # Tokenizar la entrada
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

    # Generar texto con el modelo
    output = model.generate(
        input_ids,
        max_length=max_length,
        num_return_sequences=num_return_sequences,
        no_repeat_ngram_size=2,
        repetition_penalty=1.5,
        top_p=0.95,
        temperature=0.7,
        do_sample=True,
        top_k=50,
    )

    # Decodificar y devolver la respuesta
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response

# Ejemplo de uso del chatbot
while True:
    prompt = input("You: ")
    if prompt.lower() in ["exit", "quit", "salir"]:
        break

    response = generate_response(prompt)
    print(f"Chatbot: {response}")
```

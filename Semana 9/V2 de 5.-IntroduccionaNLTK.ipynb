```python
data = [
    ("I love this movie", "positive"),
    ("This movie is terrible", "negative"),
    ("This movie is great", "positive"),
    ("I dislike this movie", "negative"),
    ("This film is amazing", "positive"),
    ("I can't stand watching this movie", "negative"),
    ("The acting in this movie is phenomenal", "positive"),
    ("I regret wasting my time on this film", "negative"),
    ("I thoroughly enjoyed this movie", "positive"),
    ("This movie lacks depth and substance", "negative"),
    ("The plot of this movie was captivating", "positive"),
    ("I found the characters in this film to be very engaging", "positive"),
    ("The special effects in this movie were impressive", "positive"),
    ("The storyline was predictable and unoriginal", "negative"),
    ("I was disappointed by the lack of character development", "negative"),
    ("The cinematography in this film was stunning", "positive"),
    ("The dialogue felt forced and unnatural", "negative"),
    ("The pacing of the movie was too slow for my liking", "negative"),
    ("I was pleasantly surprised by how much I enjoyed this film", "positive"),
    ("The ending left me feeling unsatisfied and confused", "negative"),
    ("This movie exceeded my expectations", "positive"),
    ("The performances by the actors were lackluster", "negative"),
    ("The soundtrack of this movie was mesmerizing", "positive"),
    ("The film was a complete waste of time", "negative"),
    ("I was blown away by the visuals", "positive"),
    ("The plot was convoluted and hard to follow", "negative"),
    ("I couldn't stop laughing throughout the film", "positive"),
    ("The movie was a bit too long for my taste", "negative"),
    ("The characters were well-developed and relatable", "positive"),
    ("The script was poorly written and full of clichés", "negative"),
    ("The direction of the film was masterful", "positive"),
    ("The movie failed to hold my interest", "negative"),
    ("The action scenes were thrilling and well-executed", "positive"),
    ("The film had too many plot holes", "negative"),
    ("The movie was an emotional rollercoaster", "positive"),
    ("The lead actor's performance was underwhelming", "negative"),
    ("The film's message was inspiring and thought-provoking", "positive"),
    ("The movie dragged on and felt boring", "negative"),
    ("The cinematography and visual effects were top-notch", "positive"),
    ("The movie was just okay, nothing special", "negative"),
    ("The performances were compelling and authentic", "positive"),
    ("I found the movie to be dull and uninspired", "negative"),
    ("The film had a unique and original storyline", "positive"),
    ("The movie was riddled with clichés and predictability", "negative"),
    ("The emotional scenes were handled beautifully", "positive"),
    ("The film was overhyped and disappointing", "negative"),
    ("The chemistry between the leads was fantastic", "positive"),
    ("The movie was poorly paced and lacked excitement", "negative"),
    ("The story was heartwarming and uplifting", "positive"),
    ("The film was filled with unnecessary scenes", "negative"),
    ("The movie had a great balance of humor and drama", "positive"),
    ("The characters were flat and uninteresting", "negative"),
    ("The film was a visual masterpiece", "positive"),
    ("The movie's plot was too predictable", "negative"),
    ("The acting was superb and believable", "positive"),
    ("The film failed to deliver on its promises", "negative"),
    ("The movie was a delightful surprise", "positive"),
    ("The dialogue was cringe-worthy and unrealistic", "negative"),
    ("The film's pacing was perfect", "positive"),
    ("The movie was far too slow and boring", "negative"),
    ("The story was gripping and kept me on the edge of my seat", "positive"),
    ("The film's plot was confusing and disjointed", "negative"),
    ("The movie was a brilliant piece of storytelling", "positive"),
    ("The characters were poorly developed", "negative"),
    ("The film was a joy to watch", "positive"),
    ("The movie was filled with logical inconsistencies", "negative"),
    ("The performances were stellar", "positive"),
    ("The film's ending was a letdown", "negative"),
    ("The movie was a cinematic triumph", "positive"),
    ("The film lacked a coherent plot", "negative"),
    ("The acting was top-notch", "positive"),
    ("The movie was disappointing", "negative"),
    ("The film was a masterpiece", "positive"),
    ("The movie was mediocre", "negative"),
    ("The performances were award-worthy", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was fantastic", "positive"),
    ("The film was underwhelming", "negative"),
    ("The movie was a work of art", "positive"),
    ("The film was subpar", "negative"),
    ("The movie was a joy to behold", "positive"),
    ("The film was lackluster", "negative"),
    ("The movie was a delight", "positive"),
    ("The film was boring", "negative"),
    ("The movie was captivating", "positive"),
    ("The film was uninspired", "negative"),
    ("The movie was a triumph", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was a visual feast", "positive"),
    ("The film was tedious", "negative"),
    ("The movie was a pleasure to watch", "positive"),
    ("The film was dull", "negative"),
    ("The movie was extraordinary", "positive"),
    ("The film was forgettable", "negative"),
    ("The movie was outstanding", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was exceptional", "positive"),
    ("The film was mediocre at best", "negative"),
    ("The movie was a revelation", "positive"),
    ("The film was lackluster", "negative"),
    ("The movie was a treasure", "positive"),
    ("The film was unimpressive", "negative"),
    ("The movie was superb", "positive"),
    ("The film was unremarkable", "negative"),
    ("The movie was a gem", "positive"),
    ("The film was substandard", "negative"),
    ("The movie was brilliant", "positive"),
    ("The film was underwhelming", "negative"),
    ("The movie was a fantastic experience", "positive"),
    ("The film was forgettable", "negative"),
    ("The movie was amazing", "positive"),
    ("The film was not worth watching", "negative"),
    ("The movie was delightful", "positive"),
    ("The film was a disappointment", "negative"),
    ("The movie was a great watch", "positive"),
    ("The film was unenjoyable", "negative"),
    ("The movie was highly entertaining", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was incredible", "positive"),
    ("The film was not engaging", "negative"),
    ("The movie was a thrilling ride", "positive"),
    ("The film was a chore to sit through", "negative"),
    ("The movie was a wonderful experience", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was a spectacular film", "positive"),
    ("The film was badly written", "negative"),
    ("The movie was a breathtaking journey", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was a cinematic gem", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a marvelous film", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was a phenomenal piece of work", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was a stellar production", "positive"),
    ("The film was unwatchable", "negative"),
    ("The movie was a fantastic adventure", "positive"),
    ("The film was a bore", "negative"),
    ("The movie was a superb experience", "positive"),
    ("The film was a dud", "negative"),
    ("The movie was a joy to experience", "positive"),
    ("The film was poorly done", "negative"),
    ("The movie was an incredible journey", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was a magnificent film", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was an excellent production", "positive"),
    ("The film was poorly written", "negative"),
    ("The movie was a brilliant masterpiece", "positive"),
    ("The film was not enjoyable", "negative"),
    ("The movie was an impressive piece of work", "positive"),
    ("The film was a failure", "negative"),
    ("The movie was an outstanding film", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was a captivating story", "positive"),
    ("The film was boring", "negative"),
    ("The movie was a wonderful piece of art", "positive"),
    ("The film was underwhelming", "negative"),
    ("The movie was an exceptional film", "positive"),
    ("The film was forgettable", "negative"),
    ("The movie was a fantastic story", "positive"),
    ("The film was not worth watching", "negative"),
    ("The movie was a delightful experience", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was a thrilling story", "positive"),
    ("The film was poorly done", "negative"),
    ("The movie was an amazing production", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was an extraordinary film", "positive"),
    ("The film was unremarkable", "negative"),
    ("The movie was a breathtaking experience", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was a captivating journey", "positive"),
    ("The film was boring", "negative"),
    ("The movie was a spectacular production", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a marvelous story", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was a phenomenal film", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was a stellar story", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was a fantastic piece of art", "positive"),
    ("The film was unenjoyable", "negative"),
    ("The movie was a superb production", "positive"),
    ("The film was a bore", "negative"),
    ("The movie was a magnificent story", "positive"),
    ("The film was a dud", "negative"),
    ("The movie was a wonderful journey", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was an excellent story", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was a brilliant production", "positive"),
    ("The film was boring", "negative"),
    ("The movie was an impressive story", "positive"),
    ("The film was poorly written", "negative"),
    ("The movie was an outstanding piece of work", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was a captivating film", "positive"),
    ("The film was unwatchable", "negative"),
    ("The movie was a wonderful piece of work", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was an exceptional journey", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was a fantastic film", "positive"),
    ("The film was forgettable", "negative"),
    ("The movie was an amazing story", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was an extraordinary journey", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a spectacular film", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was a marvelous piece of art", "positive"),
    ("The film was underwhelming", "negative"),
    ("The movie was a phenomenal production", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was a stellar journey", "positive"),
    ("The film was unremarkable", "negative"),
    ("The movie was a fantastic experience", "positive"),
    ("The film was boring", "negative"),
    ("The movie was a superb story", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was a magnificent production", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was a wonderful story", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was an excellent film", "positive"),
    ("The film was poorly written", "negative"),
    ("The movie was a brilliant story", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was an impressive film", "positive"),
    ("The film was boring", "negative"),
    ("The movie was an outstanding journey", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was a captivating piece of work", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a wonderful experience", "positive"),
    ("The film was forgettable", "negative"),
    ("The movie was an exceptional story", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was a fantastic film", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was an amazing experience", "positive"),
    ("The film was poorly done", "negative"),
    ("The movie was an extraordinary story", "positive"),
    ("The film was unremarkable", "negative"),
    ("The movie was a spectacular journey", "positive"),
    ("The film was boring", "negative"),
    ("The movie was a marvelous production", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was a phenomenal story", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was a stellar film", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a fantastic piece of art", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was a superb journey", "positive"),
    ("The film was poorly written", "negative"),
    ("The movie was a magnificent film", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was a wonderful piece of art", "positive"),
    ("The film was a bore", "negative"),
    ("The movie was an excellent production", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was a brilliant experience", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was an impressive journey", "positive"),
    ("The film was boring", "negative"),
    ("The movie was an outstanding piece of work", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a captivating film", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was a wonderful story", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was an exceptional production", "positive"),
    ("The film was forgettable", "negative"),
    ("The movie was a fantastic piece of work", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was an amazing journey", "positive"),
    ("The film was unremarkable", "negative"),
    ("The movie was an extraordinary film", "positive"),
    ("The film was boring", "negative"),
    ("The movie was a spectacular production", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was a marvelous journey", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was a phenomenal piece of art", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was a stellar production", "positive"),
    ("The film was poorly written", "negative"),
    ("The movie was a fantastic experience", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was a superb story", "positive"),
    ("The film was boring", "negative"),
    ("The movie was a magnificent journey", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a wonderful production", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was an excellent film", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was a brilliant piece of art", "positive"),
    ("The film was poorly done", "negative"),
    ("The movie was an impressive experience", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was an outstanding production", "positive"),
    ("The film was unremarkable", "negative"),
    ("The movie was a captivating journey", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was a wonderful piece of work", "positive"),
    ("The film was boring", "negative"),
    ("The movie was an exceptional experience", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was a fantastic story", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was an amazing production", "positive"),
    ("The film was disappointing", "negative"),
    ("The movie was an extraordinary piece of work", "positive"),
    ("The film was a flop", "negative"),
    ("The movie was a spectacular story", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was a marvelous experience", "positive"),
    ("The film was a mess", "negative"),
    ("The movie was a phenomenal film", "positive"),
    ("The film was unremarkable", "negative"),
    ("The movie was a stellar journey", "positive"),
    ("The film was a waste of time", "negative"),
    ("The movie was a fantastic piece of work", "positive"),
    ("The film was boring", "negative"),
    ("The movie was a superb production", "positive"),
    ("The film was poorly executed", "negative"),
    ("The movie was a magnificent story", "positive"),
    ("The film was a disaster", "negative"),
    ("The movie was a wonderful journey", "positive"),
    ("The film was a letdown", "negative"),
    ("The movie was an excellent piece of work", "positive"),
    ("The film was poorly done", "negative"),
    ("The movie was a brilliant film", "positive"),
    ("The film was a flop", "negative")

]

def preprocess(text):
    tokens = word_tokenize(text)
    return {word: True for word in tokens}

featuresets=[(preprocess(text),label) for (text,label) in data]
import random

# Assuming 'featuresets' is already defined
split_ratio = 0.8  # Adjust this as needed
split_index = int(len(featuresets) * split_ratio)

train_set = featuresets[:split_index]
test_set = featuresets[split_index:]
```

```python
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Extracting labels from the data
labels = [label for _, label in data]
label_counts = Counter(labels)

# Gráfico de barras para mostrar la cantidad de datos por "positivo" y "negativo"
plt.figure(figsize=(8, 6))
sns.barplot(x=list(label_counts.keys()), y=list(label_counts.values()))
plt.title('Cantidad de datos por positivo y negativo')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
all_words = [word.lower() for text, _ in data for word in nltk.word_tokenize(text)]

# Conteo de frecuencia de cada palabra
word_freq = Counter(all_words)

# Seleccionar las palabras más comunes
most_common_words = word_freq.most_common(30)

# Gráfico de barras para mostrar la frecuencia de cada palabra
words, counts = zip(*most_common_words)
plt.figure(figsize=(10, 8))
sns.barplot(y=list(words), x=list(counts), orient='h')
plt.title('Frecuencia de las palabras más comunes')
plt.xlabel('Frecuencia')
plt.ylabel('Palabra')
plt.show()
```

```python
from wordcloud import WordCloud
# Unir todas las frases en un solo texto
text = " ".join([t[0] for t in data])

# Generar la nube de palabras
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Mostrar la nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras de las Críticas de Películas')
plt.show()
```

# Entrenamiento del clasicador:

```python
classifier=nltk.NaiveBayesClassifier.train(train_set)
```

# Evaluación del clasicador:

```python
accuracy=nltk.classify.accuracy(classifier,test_set)
print("Accuracy:",accuracy)
```

# Clasicación de un nuevo texto:

```python
new_text="This movie is amazing"
new_text_features=preprocess(new_text)
predicted_label=classifier.classify(new_text_features)
print("Prediction label:",predicted_label)
```

# Métricas

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
predicted_labels=[classifier.classify(features) for (features,_) in test_set]
true_labels = [label for (_, label) in test_set]
accuracy=accuracy_score(true_labels,predicted_labels)
precision=precision_score(true_labels, predicted_labels, average='weighted')
recall=recall_score(true_labels,predicted_labels, average='weighted')
f1=f1_score(true_labels,predicted_labels, average='weighted')
print("Accuracy:",accuracy)
print("Precision:",precision)
print("Recall:",recall)
print("F1 score:",f1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ... (Your existing code for generating predicted_labels and true_labels) ...

# Classification Report
print("Classification Report:")
print(classification_report(true_labels, predicted_labels))

# Confusion Matrix
cm = confusion_matrix(true_labels, predicted_labels)
print("\nConfusion Matrix:")
print(cm)

# Confusion Matrix Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classifier.labels(), yticklabels=classifier.labels())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# ROC Curve and AUC (Assuming binary classification for simplicity)
# If you have multi-class, you'll need to adapt this part
if len(classifier.labels()) == 2:
    # Convert labels to binary (0 and 1) if they aren't already
    binary_true_labels = [1 if label == classifier.labels()[0] else 0 for label in true_labels]
    probas = [classifier.prob_classify(features).prob(classifier.labels()[0]) for (features, _) in test_set]
    fpr, tpr, thresholds = roc_curve(binary_true_labels, probas)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()

# Error Distribution
errors = [i for i in range(len(true_labels)) if true_labels[i] != predicted_labels[i]]
print("\nIndices of Misclassified Instances:", errors)

# You can further analyze the error distribution by examining the features
# of the misclassified instances or plotting histograms of relevant features.
```

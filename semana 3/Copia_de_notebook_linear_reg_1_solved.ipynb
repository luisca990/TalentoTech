- TAREA:

Calcular métricas de desempeño del modelo: MSE, RMSE, MAE, r2, entre otros.

````python
def mean_squared_error(y_true, y_pred):
    n = len(y_true)
    mse = sum((yt - yp) ** 2 for yt, yp in zip(y_true, y_pred)) / n
    return mse

def root_mean_squared_error(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = mse ** 0.5
    return rmse

def mean_absolute_error(y_true, y_pred):
    n = len(y_true)
    mae = sum(abs(yt - yp) for yt, yp in zip(y_true, y_pred)) / n
    return mae

def r2_score(y_true, y_pred):
    mean_y_true = sum(y_true) / len(y_true)
    ss_total = sum((yt - mean_y_true) ** 2 for yt in y_true)
    ss_res = sum((yt - yp) ** 2 for yt, yp in zip(y_true, y_pred))
    r2 = 1 - (ss_res / ss_total)
    return r2


mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R2:", r2)
````

## Análisis de Métricas de Desempeño del Modelo de Regresión

Vamos a analizar las métricas MSE, RMSE, MAE y R² con los valores específicos que has proporcionado para evaluar la calidad de tu modelo de regresión.

### 1. Mean Squared Error (MSE)
- **Valor proporcionado**: 3682.5912111612733
- **Significado**:
  - El MSE mide la media de los errores al cuadrado entre los valores verdaderos (y_true) y los valores predichos (y_pred).
  - Este valor elevado sugiere que el modelo tiene errores grandes cuando hace predicciones.
  - Cuanto menor sea el MSE, mejor será el ajuste del modelo a los datos.

### 2. Root Mean Squared Error (RMSE)
- **Valor proporcionado**: 60.684357219643296
- **Significado**:
  - El RMSE es la raíz cuadrada del MSE y está en las mismas unidades que la variable de respuesta.
  - Un RMSE de aproximadamente 60.68 indica que, en promedio, las predicciones del modelo están desviadas del valor real por alrededor de 60.68 unidades.
  - Un RMSE más bajo sugiere un mejor desempeño del modelo. En este caso, el valor es considerablemente alto, lo que sugiere que el modelo no predice con precisión.

### 3. Mean Absolute Error (MAE)
- **Valor proporcionado**: 50.609380582115946
- **Significado**:
  - El MAE mide la media de los errores absolutos entre los valores verdaderos y los valores predichos.
  - Un MAE de aproximadamente 50.61 indica que, en promedio, las predicciones del modelo están desviadas del valor real por alrededor de 50.61 unidades.
  - El MAE no penaliza los errores grandes tan fuertemente como el MSE o el RMSE, pero en este caso, sigue siendo bastante elevado, lo que indica que el modelo tiene un error significativo en sus predicciones.

### 4. Coeficiente de Determinación (R²)
- **Valor proporcionado**: 0.35543599211414334
- **Significado**:
  - El R² indica la proporción de la varianza en la variable dependiente que es explicable a partir de las variables independientes.
  - Un R² de aproximadamente 0.355 sugiere que solo el 35.5% de la variabilidad en los datos es explicada por el modelo. Esto es relativamente bajo, indicando que el modelo no captura bien la relación entre las variables independientes y la variable dependiente.
  - Un R² cercano a 1 indica un buen ajuste del modelo a los datos, mientras que un R² cercano a 0 indica un ajuste pobre.

# Importar librerias

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Procesamiento de datos

```python
time_series = np.array([i for i in range(1,16)])
print(time_series)
train_size=int(len(time_series)*0.6)
train,test=time_series[:train_size],time_series[train_size:]
print(train)
print(test)
```

# Preparación de datos para LSTM

```python
def create_dataset(dataset,seq_lenght):
    data_x,data_y=[],[]
    for i in range(len(dataset)-seq_lenght):
        data_x.append(dataset[i:i+seq_lenght])
        data_y.append(dataset[i+seq_lenght])
    return np.array(data_x),np.array(data_y)

seq_lenght=4
train_x,train_y=create_dataset(train,seq_lenght)
test_x,test_y=create_dataset(test,seq_lenght)
print(train_x)
print(train_y)
```

# Modelo

```python
model=Sequential()
model.add(LSTM(16, activation='relu',input_shape=(seq_lenght,1)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')
model.fit(train_x,train_y,epochs=500,batch_size=1,verbose=2)
```

# Evaluación del modelo

```python
Evaluacion=model.evaluate(test_x,test_y)
print(Evaluacion)
```

```python
y_pred=model.predict(test_x)
print(y_pred)
```

# Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error
from math import sqrt
```

### Generar datos de ejemplo

```python
np.random.seed(0)
dates=pd.date_range(start='2022-01-01',end='2022-12-31')
values=np.arange(len(dates))*0.2+np.random.randn(len(dates))*50
ts=pd.Series(values,index=dates)
```

### Dividir los datos train y test

```python
train_data=ts[:int(len(ts)*0.8)]
test_data=ts[int(len(ts)*0.8):]
```

### Modelo ARIMA

```python
model=ARIMA(train_data,order=(10,1,10))
model_fit=model.fit()
```

### Pronóstico

```python
forecast=model_fit.forecast(steps=len(test_data))
```

### Gráfica de los datos originales y el pronóstico

```python
plt.figure(figsize=(20,6))
plt.plot(train_data,label='Datos de entrenamiento')
plt.plot(test_data,label='Datos de prueba', color='green')
plt.plot(test_data.index,forecast,label='Pronóstico', color='red')
plt.legend()
plt.title('Pronóstico de la serie de tiempo')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

# Métricas

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
mae = mean_absolute_error(test_data, forecast)
rmse = np.sqrt(mean_squared_error(test_data, forecast))
mape = np.mean(np.abs((test_data - forecast) / test_data)) * 100
r2 = r2_score(test_data, forecast)

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}%')
print(f'R²: {r2}')

# Cálculo de los errores (residuos)
errors = test_data - forecast

# Gráfico de la distribución de los errores
plt.figure(figsize=(8, 6))
plt.hist(errors, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribución de los errores')
plt.xlabel('Error')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Listas x y h
x = [0, 1, 2, 3, 4, 5, 0]
h = [0, -1, 5, 3, -2, 1, 0]

# Realizar la convolución
y = np.convolve(x, h, mode='full')

# Mostrar la operación de convolución
print("Resultado de la convolución:")
print(y)

# Graficar el resultado
plt.stem(y, use_line_collection=True)
plt.title("Resultado de la convolución de x y h")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Listas x y h
x = [0, 1, 2, 3, 4, 5, 0]
h = [0, -1, 5, 3, -2, 1, 0]

# Calcular el desplazamiento necesario
desplazamiento = 1 - 2

# Desplazar la señal h
h_desplazada = np.roll(h, desplazamiento)

# Realizar la convolución
y = np.convolve(x, h_desplazada, mode='full')

# Mostrar la operación de convolución
print("Resultado de la convolución:")
print(y)

# Graficar el resultado
plt.stem(y, use_line_collection=True)
plt.title("Resultado de la convolución de x y h (desplazada)")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Listas x y h
x = [0, 1, 2, 3, 4, 5, 0]
h = [0, -1, 5, 3, -2, 1, 0]

# Desplazar la señal h de modo que el 3 de h esté en el índice 1 de x
h_desplazada = [5, 3, -2, 1, 0, 0, 0]

# Realizar la convolución manualmente
y = np.zeros(len(x) + len(h_desplazada) - 1)
for i in range(len(x)):
    for j in range(len(h_desplazada)):
        y[i + j] += x[i] * h_desplazada[j]

# Mostrar la operación de convolución
print("Resultado de la convolución:")
print(y)

# Graficar el resultado
plt.stem(y, use_line_collection=True)
plt.title("Resultado de la convolución de x y h (desplazada)")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
```

```python
import cv2
```

```python
image= cv2.imread("descarga.jpg")
```

```python
cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Rotaci√≥n

```python
height,width=image.shape[:2]
center=(width/2,height/2)
angle=45
M=cv2.getRotationMatrix2D(center,angle,1)
rotated=cv2.warpAffine(image,M,(width,height))
cv2.imshow("rotated",rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Traslacion

```python
new_width,new_height=200,300
scaled=cv2.resize(image,(new_width,new_height))
cv2.imshow("scaled",scaled)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Suavizado

```python
smoothed=cv2.GaussianBlur(image,(7,7),0)
cv2.imshow("smoothed",smoothed)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Realce

```python
import numpy as np
kernel=np.array([[-1,-1,-1]
                 ,[-1,9,-1]
                 ,[-1,-1,-1]])
sharpened=cv2.filter2D(image,-1,kernel)
cv2.imshow("sharpened",sharpened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Detecci√≥n de bordes

```python
image2=cv2.imread("descarga.jpg", cv2.IMREAD_GRAYSCALE)
sobelx=cv2.Sobel(image2,cv2.CV_64F,1,0,ksize=5)
sobely=cv2.Sobel(image2,cv2.CV_64F,0,1,ksize=5)
edges=cv2.magnitude(sobelx,sobely)
edges=cv2.normalize(edges,None,0,255,cv2.NORM_MINMAX,dtype=cv2.CV_8U)
cv2.imshow("edges",edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# prompt: - Carga una imagen en color y en escala de grises utilizando OpenCV.
# - Muestra ambas im√°genes en ventanas separadas.
# - Carga una imagen y rota la imagen 90 grados en sentido horario.
# - Escala la imagen al doble de su tama√±o original.
# - Muestra ambas im√°genes resultantes

import cv2

# Carga una imagen en color y en escala de grises
imagen_color = cv2.imread("descarga.jpg")
imagen_grises = cv2.imread("descarga.jpg", cv2.IMREAD_GRAYSCALE)

# Muestra ambas im√°genes en ventanas separadas
cv2.imshow("Imagen en color", imagen_color)
cv2.imshow("Imagen en escala de grises", imagen_grises)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Carga una imagen y rota la imagen 90 grados en sentido horario
imagen = cv2.imread("descarga.jpg")
(h, w) = imagen.shape[:2]
centro = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(centro, -90, 1.0)
imagen_rotada = cv2.warpAffine(imagen, M, (w, h))

# Escala la imagen al doble de su tama√±o original
imagen_escalada = cv2.resize(imagen, None, fx=2, fy=2)

# Muestra ambas im√°genes resultantes
cv2.imshow("Imagen rotada", imagen_rotada)
cv2.imshow("Imagen escalada", imagen_escalada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# prompt: - Carga una imagen en escala de grises.
# - Aplica un fltro Gaussiano para suavizar la imagen.
# - Aplica un filtro de realce para mejorar los detalles de la imagen.
# - Muestra ambas im√°genes resultantes.
# - Carga una imagen en escala de grises.
# - Utiliza el operador Sobel para detectar bordes en la imagen.
# - Muestra la imagen original y la imagen con los bordes detectados

# Carga una imagen en escala de grises.
imagen_grises = cv2.imread("descarga.jpg", cv2.IMREAD_GRAYSCALE)

# Aplica un filtro Gaussiano para suavizar la imagen.
imagen_suavizada = cv2.GaussianBlur(imagen_grises, (5, 5), 0)

# Aplica un filtro de realce para mejorar los detalles de la imagen.
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
imagen_realzada = cv2.filter2D(imagen_suavizada, -1, kernel)

# Muestra ambas im√°genes resultantes.
cv2.imshow("Imagen suavizada", imagen_suavizada)
cv2.imshow("Imagen realzada", imagen_realzada)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Carga una imagen en escala de grises.
imagen_grises = cv2.imread("descarga.jpg", cv2.IMREAD_GRAYSCALE)

# Utiliza el operador Sobel para detectar bordes en la imagen.
sobelx = cv2.Sobel(imagen_grises, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(imagen_grises, cv2.CV_64F, 0, 1, ksize=5)
bordes = cv2.magnitude(sobelx, sobely)
bordes = cv2.normalize(bordes, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

# Muestra la imagen original y la imagen con los bordes detectados
cv2.imshow("Imagen original", imagen_grises)
cv2.imshow("Bordes detectados", bordes)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Ejemplo académico:

Definiendo funciones en python para calcular:

 W x + b = Z --> relu ---> y

````python
import numpy as np

def get_output_from_MLP_input(W, x, b, activation="linear"):
    print(f"W = \n{W}\n")
    print("x = ", end='\n')
    print(x)
    print("b = ", end='\n')
    print(b)

    # Operaciones
    # Z = W x + b
    Z = np.dot(W, x) + b

    print("Z = ", end='\n')
    print(Z)

    # Funciones de Activación
    if activation == "linear":
        y = Z

    elif activation == "relu":
        y = np.maximum(0, Z)

    elif activation == "sigmoid":
        y = 1 / (1 + np.exp(-Z))

    elif activation == "tanh":
        y = np.tanh(Z)

    elif activation == "softmax":
        exp_Z = np.exp(Z - np.max(Z))  # Estabilidad numérica
        y = exp_Z / np.sum(exp_Z)

    elif activation == "leaky_relu":
        alpha = 0.01
        y = np.where(Z > 0, Z, alpha * Z)

    elif activation == "swish":
        beta = 1  # parámetro de swish, ajustable
        y = Z / (1 + np.exp(-beta * Z))

    elif activation == "elu":
        alpha = 1.0
        y = np.where(Z > 0, Z, alpha * (np.exp(Z) - 1))

    else:
        raise ValueError(f"Función de activación '{activation}' no reconocida.")

    print("y = ", end='\n')
    print(y)

    return y

# Ejemplo de uso
W = np.array([[0.2, 0.4], [0.4, 0.6], [0.6, 0.8]])
x = np.array([0.5, 0.9])
b = np.array([0.1, 0.2, 0.3])
output = get_output_from_MLP_input(W, x, b, activation="relu")
````

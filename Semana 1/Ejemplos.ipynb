# Introducción a Python

En este notebook, exploro algunos conceptos básicos de Python, incluyendo cálculos, visualización de datos y la creación de funciones.

## Cálculos básicos

A continuación, haré algunos ejemplos.

```python
# Suma
a = 5
b = 3
suma = a + b
print(f"La suma de {a} y {b} es {suma}")

# Resta
resta = a - b
print(f"La resta de {a} y {b} es {resta}")

# Multiplicación
multiplicacion = a * b
print(f"La multiplicación de {a} y {b} es {multiplicacion}")

# División
division = a / b
print(f"La división de {a} entre {b} es {division}")
```

## Visualización de datos

La visualización de datos es una parte crucial del análisis de datos. En Python, hay bibliotecas como Matplotlib y Seaborn para crear gráficos de manera sencilla.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Gráfico de línea con Matplotlib
plt.plot(x, y, marker='o')
plt.title('Gráfico de Línea')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Gráfico de dispersión con Seaborn
sns.scatterplot(x=x, y=y)
plt.title('Gráfico de Dispersión')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Funciones en Python

Las funciones permiten reutilizar código y hacer programas más organizados. A continuación, se crea una función simple para sumar dos números.

```python
def sumar(a, b):
    """Esta función suma dos números y devuelve el resultado."""
    return a + b

# Uso de la función
resultado = sumar(10, 5)
print(f"El resultado de la suma es {resultado}")
```

## Ejercicio práctico

Está función recibe una lista de números y devuelve un diccionario con el valor mínimo, máximo y promedio de los números en la lista. Luego, visualiza estos valores utilizando un gráfico de barras.

```python
def calcular_estadisticas(numeros):
    """Esta función calcula el valor mínimo, máximo y promedio de una lista de números."""
    min_val = min(numeros)
    max_val = max(numeros)
    avg_val = sum(numeros) / len(numeros)
    return {'min': min_val, 'max': max_val, 'avg': avg_val}

# Lista de ejemplo
numeros = [2, 4, 6, 8, 10, 12, 14]

# Cálculo de estadísticas
estadisticas = calcular_estadisticas(numeros)
print(estadisticas)

# Visualización de estadísticas
labels = list(estadisticas.keys())
values = list(estadisticas.values())

plt.bar(labels, values)
plt.title('Estadísticas de Números')
plt.xlabel('Medida')
plt.ylabel('Valor')
plt.show()
```

## Importación de Plotly

Primero, importar las bibliotecas necesarias para crear los gráficos.

```python
import plotly.graph_objs as go
import plotly.express as px

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

## Gráfico de Dispersión Interactivo

Se usa Plotly Express para crear un gráfico de dispersión interactivo.

```python
# Crear gráfico de dispersión
fig = px.scatter(x=x, y=y, title='Gráfico de Dispersión Interactivo', labels={'x':'X', 'y':'Y'})
fig.show()
```

## Gráfico de Barras Interactivo

Ahora, se crea un gráfico de barras interactivo utilizando Plotly Graph Objects.

```python
# Datos de ejemplo para el gráfico de barras
estadisticas = {'min': 2, 'max': 11, 'avg': 5.6}
labels = list(estadisticas.keys())
values = list(estadisticas.values())

# Crear gráfico de barras
fig = go.Figure(data=[go.Bar(x=labels, y=values)])
fig.update_layout(title='Gráfico de Barras Interactivo', xaxis_title='Medida', yaxis_title='Valor')
fig.show()
```

# Ejemplo del uso de un Gauge

```python
import plotly.graph_objects as go

# Crear el gráfico de tipo gauge
fig = go.Figure(go.Indicator(
    mode = "gauge+number+delta",
    value = 270,
    delta = {'reference': 200, 'increasing': {'color': "green"}},
    gauge = {
        'axis': {'range': [0, 500]},
        'bar': {'color': "darkblue"},
        'steps' : [
            {'range': [0, 250], 'color': "lightgray"},
            {'range': [250, 400], 'color': "gray"}],
        'threshold' : {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': 490}},
    title = {'text': "Speedometer"}))

fig.show()
```

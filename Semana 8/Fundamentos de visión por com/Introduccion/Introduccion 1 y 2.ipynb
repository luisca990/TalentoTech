```python
image_path='/content/descarga.jpg'
Image=skimage.io.imread(image_path)
print(type(Image))
print(Image.shape)
skimage.io.imshow(Image)
skimage.io.show()
```

```python
rotated_image=skimage.transform.rotate(Image,45)
skimage.io.imshow(rotated_image)
skimage.io.show
```

```python
smoothed_image=skimage.filters.gaussian(Image,sigma=1)
skimage.io.imshow(smoothed_image)
skimage.io.show
```

```python
gray_image=skimage.color.rgb2gray(Image)
skimage.io.imshow(gray_image)
skimage.io.show
```

```python
rotated_image90=skimage.transform.rotate(Image,90)
skimage.io.imshow(rotated_image90)
skimage.io.show
```

```python
scaled_image=skimage.transform.rescale(Image,scale=1.5)
skimage.io.imshow(scaled_image)
skimage.io.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

imagen_reflejada = np.fliplr(Image)

# Mostrar la Image original y la Image reflejada
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(Image)
axes[0].set_title('Image Original')
axes[0].axis('off')

axes[1].imshow(imagen_reflejada)
axes[1].set_title('Image Reflejada Horizontalmente')
axes[1].axis('off')

plt.show()
```

```python
gaussian_image = skimage.filters.gaussian(Image, sigma=2)
skimage.io.imshow(gaussian_image)
skimage.io.show
```

```python
sobel_image = skimage.filters.sobel(Image)
skimage.io.imshow(sobel_image)
skimage.io.show
```

```python
# prompt: Utiliza algún método de segmentación de umbral para separar objetos de interés del fondo en la imagen.

from skimage.filters import threshold_otsu

# Convertir la imagen a escala de grises si no lo está
gray_image = skimage.color.rgb2gray(Image)

# Calcular el umbral óptimo utilizando el método de Otsu
thresh = threshold_otsu(gray_image)

# Crear una máscara binaria
binary_mask = gray_image > thresh

# Mostrar la máscara binaria
skimage.io.imshow(binary_mask)
skimage.io.show()
```

```python
# prompt: Aplica operaciones morfológicas como erosión, dilatación, apertura y cierre a la imagen.

from skimage.morphology import erosion, dilation, opening, closing, disk

# Crear un elemento estructurante (por ejemplo, un disco)
selem = disk(6)

# Erosión
eroded_image = erosion(binary_mask, selem)
skimage.io.imshow(eroded_image)
skimage.io.show()

# Dilatación
dilated_image = dilation(binary_mask, selem)
skimage.io.imshow(dilated_image)
skimage.io.show()

# Apertura
opened_image = opening(binary_mask, selem)
skimage.io.imshow(opened_image)
skimage.io.show()

# Cierre
closed_image = closing(binary_mask, selem)
skimage.io.imshow(closed_image)
skimage.io.show()
```
